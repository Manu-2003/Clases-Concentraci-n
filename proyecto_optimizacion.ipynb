```python
# Crear un diccionario para almacenar el conteo de cada tipo de transmisión
transmission_count = {}

# Iterar sobre cada fila en el DataFrame
for transmission in data['transmission']:
    # Si el valor de transmisión no está en el diccionario, inicializarlo a 1
    if transmission not in transmission_count:
        transmission_count[transmission] = 1
    # Si ya existe, incrementar el contador en 1
    else:
        transmission_count[transmission] += 1

# Mostrar el conteo de cada tipo de transmisión
for transmission, count in transmission_count.items():
    print(f"Transmisión: {transmission}, Count: {count}")
```

```python
# Encontrar el valor más frecuente
most_frequent = data['transmission'].mode()[0]
# Rellenar los NaN con el valor más frecuente
data['transmission'].fillna(most_frequent, inplace=True)
```

```python
# Lista de transmisiones permitidas
allowed_transmissions = ['manual', 'automatic']

# Reemplazar valores no permitidos con 'automatic'
data['transmission'] = data['transmission'].apply(lambda x: x if x in allowed_transmissions else 'automatic')

# Mostrar los valores únicos para confirmar
print(data['transmission'].unique())
print(data['transmission'].nunique())
```

```python
print(data.make.unique())
print(data.make.nunique())
```

```python
# Asegurarse de que las columnas numéricas contengan solo valores numéricos
data['odometer'] = pd.to_numeric(data['odometer'], errors='coerce')
data['mmr'] = pd.to_numeric(data['mmr'], errors='coerce')
data['sellingprice'] = pd.to_numeric(data['sellingprice'], errors='coerce')

# Imputar valores numéricos con la mediana
data['odometer'].fillna(data['odometer'].median(), inplace=True)
data['mmr'].fillna(data['mmr'].median(), inplace=True)
data['sellingprice'].fillna(data['sellingprice'].median(), inplace=True)

# Imputar valores categóricos con la moda
data['make'].fillna(data['make'].mode()[0], inplace=True)
data['model'].fillna(data['model'].mode()[0], inplace=True)
data['body'].fillna(data['body'].mode()[0], inplace=True)
data['color'].fillna(data['color'].mode()[0], inplace=True)
data['interior'].fillna(data['interior'].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

```python
# Rellenar valores nulos en 'seller' con la moda
seller_mode = data['seller'].mode()[0]
data['seller'].fillna(seller_mode, inplace=True)
```

```python
print(data.make.unique())
print(data.make.nunique())
```

```python
# Convertir todas las palabras en minúsculas y capitalizar la primera letra de cada palabra
data['make'] = data['make'].str.title()

# Estandarizar nombres comunes
data['make'].replace({'Bmw': 'BMW', 'Chev Truck': 'Chevrolet', 'Ford Tk': 'Ford', 'Gmc Truck': 'GMC', 'Hyundai Tk': 'Hyundai', 'Mazda Tk': 'Mazda', 'Oldsmobile': 'Oldsmobile', 'Plymouth': 'Plymouth', 'Rolls-Royce': 'Rolls Royce', 'Mercedes-B': 'Mercedes-Benz'}, inplace=True)
```

```python
colors_validos = ['white', 'gray', 'black', 'red', 'silver', 'blue', 'brown', 'beige', 'purple',
                  'burgundy', 'gold', 'yellow', 'green', 'charcoal', 'orange', 'off-white',
                  'turquoise', 'pink', 'lime']

# Eliminar valores no válidos y convertir a minúsculas
data['color'] = data['color'].apply(lambda x: x.lower() if x in colors_validos else np.nan)

# Eliminar duplicados y valores nulos
data['color'].dropna(inplace=True)
data['color'] = data['color'].str.title()  # Capitalizar la primera letra de cada palabra

# Si hay valores duplicados, elimínalos
data['color'].drop_duplicates(inplace=True)

# Mostrar los colores únicos después del proceso de limpieza
print(data['color'].unique())
```

```python
# Reemplazar los valores nulos en 'color' con la moda
moda_color = data['color'].mode()[0]
data['color'].fillna(moda_color, inplace=True)
```
